1 Read the earthquakes.csv file and save its content into a variable. Then use bar chart to visualize magnitude values being 8.2 or over so that date values will be presented in x-axle and magnitude in y-axle.

```python
import pandas as pd

df = pd.read_csv('data_files/earthquakes.csv', parse_dates=[['Date', 'Time']])

df_filtered = df[df['Magnitude'] >= 8.2]

ax = df_filtered.plot(kind='bar', x='Date_Time', y='Magnitude', color='blue', legend=None, rot=85)

ax.set_title('Magnitude of Earthquakes')
ax.set_xlabel('Date')
ax.set_ylabel('Magnitude');
```

2 Read the gamedata.json file to a new variable. Visualize the data with pie chart so that share count will be counted for each Steam rating (in other words how many percentage out of all games have received which rating).

```python
import pandas as pd

df = pd.read_json('data_files/gamedata.json')

rating_counts = df['steamRatingText'].value_counts()

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
meanings = list(rating_counts.index)

ax = rating_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, title='Steam Ratings', colors=colors, labels=None, pctdistance=1.1)
ax.legend(labels=meanings, bbox_to_anchor=(1, 0.5));
```

3 Use the same game_data.json as above and create a scatter plot where you compare the relationships for metacriticScore and salePrice.

```python
import pandas as pd

df = pd.read_json('data_files/gamedata.json')

ax = df.plot(kind='scatter', x='metacriticScore', y='salePrice', alpha=1)

ax.set_title('Metacritic Score vs Sale Price')
ax.set_xlabel('Metacritic Score')
ax.set_ylabel('Sale Price');
```
